```python
import matplotlib.pyplot as plt
```

```python
x=["science","commerce","arts"]
h=[200,300,500]
c=["red", "green", "orange"]
plt.bar(x,h, width=0.5, color=c)
plt.xlabel("courses")
plt.ylabel("enrolled students")
plt.title("students admission in 2020")
plt.show()
```

```python
y=["science","commerce","arts"]
w=[200,300,500]
c=["red", "green", "orange"]
plt.barh(y,w,0.5, color=c)
plt.xlabel("courses")
plt.ylabel("enrolled students")
plt.title("students admission in 2020")
plt.show()
```

```python
y=["science","commerce","arts"]
w=[200,300,500]
c=["red", "green", "orange"]
plt.barh(y,w,[0.5, 0.4, 0.3], color=c)
plt.xlabel("courses")
plt.ylabel("enrolled students")
plt.title("students admission in 2020")
plt.show()
```
