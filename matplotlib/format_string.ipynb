```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40])
plt.plot([5, 8, 12, 15])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
```

```python
plt.plot([0, 20, 30, 40], [5, 8, 12, 15], "r+-")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
```

```python
plt.plot([0, 20, 30, 40], [5, 8, 12, 15], "k.:")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
```
