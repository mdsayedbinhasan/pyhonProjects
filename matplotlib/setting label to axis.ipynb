```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.show()
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label",{"size":20, "color":"blue"})
plt.ylabel("y label")
plt.title("my plot")
plt.show()
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label",{"size":20, "color":"blue"}, labelpad=20)
plt.ylabel("y label")
plt.title("my plot")
plt.show()
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label",{"size":20, "color":"blue"}, labelpad=20)
plt.ylabel("y label", {"size":20, "color":"blue"}, labelpad=20)
plt.title("my plot")
plt.show()
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label",{"size":20, "color":"blue"}, labelpad=20, backgroundcolor="orange")
plt.ylabel("y label", {"size":20, "color":"blue", "backgroundcolor":"cyan"}, labelpad=20)
plt.title("my plot")
plt.show()
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label",{"size":20, "color":"blue"}, labelpad=20, backgroundcolor="orange")
plt.ylabel("y label", {"size":20, "color":"blue", "backgroundcolor":"cyan"}, labelpad=20)
plt.title("my plot")
plt.show(block=True)
print("hello")
```

```python
plt.plot([10,20,30,40])
plt.xlabel("x label",{"size":20, "color":"blue"}, labelpad=20, backgroundcolor="orange")
plt.ylabel("y label", {"size":20, "color":"blue", "backgroundcolor":"cyan"}, labelpad=20)
plt.title("my plot")
plt.show(block=False)
print("hello")
```
