```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4],[5,6,7,8], alpha=0.5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
```

```python
plt.plot([10,20,30,40], linewidth=20)
plt.plot([10,20,60,80], linewidth=20)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
```

```python
plt.plot([10,20,30,40], linewidth=20, zorder=2)
plt.plot([10,20,60,80], linewidth=20, zorder=1)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("first plot")
plt.show()
```
