```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.tick_params(length=10)
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.minorticks_on()
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.tick_params(direction="inout", length=10, width=5, pad=20, color="red")
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.minorticks_on()
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.tick_params(labelsize=20, colors="red")
plt.show()
```
