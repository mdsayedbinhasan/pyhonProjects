```python
t=np.array([1,2,3,4])
l1,=plt.plot(t**2,t,color="red",label="label1")
l2,=plt.plot(t**3,t,color="green",label="label2")
l3,=plt.plot(t,t**2,color="cyan",label="label3")
l4,=plt.plot(t,t**3,color="blue",label="label4")

plt.legend((l1,l2), ["label1", "label2"], loc="lower right")
plt.legend((l3,l4), ["label3", "label4"])

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("legend function")
plt.grid()
plt.show()
```

```python
t=np.array([1,2,3,4])
l1,=plt.plot(t**2,t,color="red",label="label1")
l2,=plt.plot(t**3,t,color="green",label="label2")
l3,=plt.plot(t,t**2,color="cyan",label="label3")
l4,=plt.plot(t,t**3,color="blue",label="label4")

legend1=plt.legend((l1,l2), ["label1", "label2"], loc="lower right")
plt.gca().add_artist(legend1)
plt.legend((l3,l4), ["label3", "label4"])

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("legend function")
plt.grid()
plt.show()
```
