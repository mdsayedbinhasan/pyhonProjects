```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40], marker="o")
plt.show()
```

```python
plt.plot([10,20,30,40], marker="o")
plt.annotate("plot", (0,10))
plt.show()
```

```python
plt.plot([10,20,30,40], marker="o")
plt.annotate("plot", (0,10), (0.5,25))
plt.show()
```

```python
plt.plot([10,20,30,40], marker="o")
plt.annotate("plot", (1,20), (0.5,25), arrowprops={})
plt.show()
```
