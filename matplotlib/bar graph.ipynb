```python
import matplotlib.pyplot as plt
```

```python
x=["science","commerce","arts"]
h=[200,300,500]
plt.bar(x,h)
plt.xlabel("courses")
plt.ylabel("enrolled students")
plt.title("students admission in 2020")
plt.show()
```

```python
x=["science","commerce","arts"]
h=[200,300,500]
c=["red", "green", "blue"]
plt.bar(x,h, color=c)
plt.xlabel("courses")
plt.ylabel("enrolled students")
plt.title("students admission in 2020")
plt.show()
```
