```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40], linewidth = 20)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], antialiased=True, linewidth = 12)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], "g", antialiased=True, linewidth = 12)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], "g", antialiased=True, linewidth = 12, dashes=[2, 1])
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], linestyle="steps")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], marker="+", color="g")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], marker="+", markeredgecolor="r")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], marker=".", color="g", markersize=10)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40], marker=".", markerfacecolor="r", markersize=20)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
plt.plot([10, 20, 30, 40, 50, 60, 70, 80], marker=".", markerfacecolor="r", markersize=20, markevery=2)
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```
