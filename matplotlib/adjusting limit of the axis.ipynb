```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.show()
```

```python
plt.plot([1,2,3,4])
left, right = plt.xlim()
print(left, right)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.xlim(0, 5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.show()
```

```python
plt.plot([1,2,3,4])
plt.xlim(0, 5)
plt.ylim(2,5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.show()
```
