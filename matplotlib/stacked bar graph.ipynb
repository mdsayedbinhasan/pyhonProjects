```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["usa", "uk", "russia", "bd", "china"]
medals=[800,700,600,650,450]
plt.bar(x,medals,0.4)
plt.xlabel("countries")
plt.ylabel("medals")
plt.title("medals vs countries")
plt.show()
```

```python
x=["usa", "uk", "russia", "bd", "china"]
gold=[10,20,30,40,50]
silver=[15,25,35,45,55]
bronze=[12,24,36,48,60,72]
plt.bar(x,gold,0.4, label="gold")
plt.bar(x,silver,0.4, label="silver")
plt.bar(x,bronze,0.4, label="bronze")
plt.xlabel("countries")
plt.ylabel("medals")
plt.title("medals vs countries")
plt.show()
```
