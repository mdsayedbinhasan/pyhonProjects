```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t = np.array([1,2,3,4])
plt.plot(t**2, t, color="red")
plt.plot(t**3, t, color="green")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.show()
```

```python
t = np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="squares")
plt.plot(t**3, t, color="green", label="cubes")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.legend()
plt.show()
```

```python
t = np.array([1,2,3,4])
plt.plot(t**2, t, color="red",)
plt.plot(t**3, t, color="green")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.legend(["squares", "cubes"])
plt.show()
```

```python
t = np.array([1,2,3,4])
line1, = plt.plot(t**2, t, color="red",)
line2, = plt.plot(t**3, t, color="green")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.legend([line1, line2],["squares", "cubes"])
plt.show()
```

```python
t = np.array([1,2,3,4])
line1, = plt.plot(t**2, t, color="red",)
line2, = plt.plot(t**3, t, color="green")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.legend([line1, line2],["squares", "cubes"], loc="upper right")
plt.show()
```

```python
t = np.array([1,2,3,4])
line1, = plt.plot(t**2, t, color="red",)
line2, = plt.plot(t**3, t, color="green")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.legend([line1, line2],["squares", "cubes"], loc="lower right")
plt.show()
```

```python
t = np.array([1,2,3,4])
line1, = plt.plot(t**2, t, color="red",)
line2, = plt.plot(t**3, t, color="green")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.legend([line1, line2],["squares", "cubes"], ncol=2, loc="lower right")
plt.show()
```
