```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t = [0, 30, 45, 60, 90]
x = [i*(np.pi/180) for i in t]
plt.plot(x, np.sin(x), marker="+")
plt.plot(x, np.cos(x), marker=".")
plt.show()
```

```python
t = [0, 30, 45, 60, 90]
x = [i*(np.pi/180) for i in t]
plt.plot(t, np.sin(x), marker="+")
plt.plot(t, np.cos(x), marker=".")
plt.xticks(t)
plt.show()
```

```python
t = [0, 30, 45, 60, 90]
x = [i*(np.pi/180) for i in t]
plt.plot(t, np.sin(x), marker="+")
plt.plot(t, np.cos(x), marker=".")
plt.xticks(t)
plt.title("sine cosine graph")
plt.xlabel("angle")
plt.ylabel("sine cosine value")
plt.show()
```

```python
t = [0, 30, 45, 60, 90]
x = [i*(np.pi/180) for i in t]
plt.plot(t, np.sin(x), marker="+", label="sin")
plt.plot(t, np.cos(x), marker=".", label="cos")
plt.xticks(t)
plt.title("sine cosine graph")
plt.xlabel("angle")
plt.ylabel("sine cosine value")
plt.legend()
plt.show()
```
