```python
import matplotlib.pyplot as plt
```

```python
w=0.4
x=["cse","eee", "iec", "mech", "ece"]
boys=[40,50,35,65,44]
girls=[65, 41, 37, 57, 48]
plt.bar(x,boys,w, label="boys")
plt.bar(x,girls,w,label="girls")
plt.xlabel("course")
plt.ylabel("students")
plt.title("student vs courses")
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
w=0.4
x=["cse","eee", "iec", "mech", "ece"]
boys=[40,50,35,65,44]
girls=[65, 41, 37, 57, 48]
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
plt.bar(bar1,boys,w, label="boys")
plt.bar(bar2,girls,w,label="girls")
plt.xlabel("course")
plt.ylabel("students")
plt.title("student vs courses")
plt.legend()
plt.show()
```
