```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.show()
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.grid(b=True)
plt.show()
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.grid(b=True, color="red")
plt.show()
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.minorticks_on()
plt.grid(b=True, which="minor")
plt.show()
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.grid(b=True, axis="x")
plt.show()
```

```python
plt.plot([10,20,30],[1,2,3])
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my graph")
plt.minorticks_on()
plt.grid(b=True, which="minor", axis="x")
plt.show()
```
