```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlim(0,5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlim(0,5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.xticks(np.arange(2,14), labels=["A", "B"])
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlim(0,5)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.xticks(np.arange(2,14), labels=["A", "B"], size="xx-large", rotation=120)
plt.show()
```

```python
t=np.array([1,2,3,4])
plt.plot(t**2, t, color="red", label="label1")
plt.plot(t**3, t, color="green", label="label2")
plt.xlim(0,2)
plt.ylim(0,2)
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("my plot")
plt.grid()
plt.xticks(np.arange(2,14), labels=["A", "B"])
plt.yticks(np.arange(2,14), labels=["A", "B"])
plt.show()
```
