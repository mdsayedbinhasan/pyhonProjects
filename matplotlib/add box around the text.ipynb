```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40], marker="o")
plt.text(1,20,"point", bbox={})
plt.text(2,30,"this is the point to add")
plt.show()
```

```python
plt.plot([10,20,30,40], marker="o")
plt.text(1,20,"point")
plt.annotate("(1,20)", (1,20), (.5, 25), arrowprops={}, bbox={"boxstyle":"circle"})
plt.text(2,30,"this is the point to add")
plt.show()
```
