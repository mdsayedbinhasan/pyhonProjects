```python
gwarming=pd.read_csv("Who_is_responsible_for_global_warming.csv")
gwarming
```

```python
gwarming.drop(columns=["Country Code", "Indicator Name", "Indicator Code"])
```

```python
gwarming=gwarming.drop(columns=["Country Code", "Indicator Name", "Indicator Code"],axis=1).set_index("Country Name")
gwarming.head(6)
```

```python
sns.heatmap(gwarming)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(gwarming)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(gwarming, vmin=0,vmax=20)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(gwarming, vmin=0,vmax=20, cmap="Accent")
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(gwarming, vmin=0,vmax=20, cmap="Accent", annot=True)
```

```python
plt.figure(figsize=(16,9))
annot_kws={"fontsize":10,
          "fontstyle":"italic",
          "color":"k",
          "alpha":0.6,
          "verticalalignment":"center",
          "backgroundcolor":"w"}
sns.heatmap(gwarming, vmin=0,vmax=20, cmap="Accent", annot=True, annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(gwarming, vmin=0,vmax=20, cmap="Accent", annot=True, linewidths=4)
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(gwarming)
ax.set(title="Who_is_responsible_for_global_warming",
      xlabel="CO2 emission",
      ylabel="country name")
sns.set(font_scale=5)
```
