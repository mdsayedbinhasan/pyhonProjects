```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
gwarming=pd.read_csv("Who_is_responsible_for_global_warming.csv")
gwarming
```

```python
gwarming.drop(columns=["Country Code", "Indicator Name", "Indicator Code"])
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(gwarming.corr(), annot=True, linewidths=4)
ax.tick_params(size=10, color="w", labelsize=10, labelcolor="w")
plt.title("Who_is_responsible_for_global_warming", fontsize=25)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                      columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(), annot=True, linewidths=4)
ax.tick_params(size=10, color="w", labelsize=10, labelcolor="w")
plt.title("breast cancer", fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(cancer_df.corr(), annot=True, linewidth=4)
ax.tick_params(size=10, color="w", labelsize=10, labelcolor="w")
plt.title("breast cancer", fontsize=25)
plt.show()
```
