```python
import pandas as pd
```

```python
sr1=pd.Series([2,3,4,5,6])
sr1
```

```python
sr2=pd.Series([1,7,8,9,10])
sr2
```

```python
pd.concat([sr1, sr2])
```

```python
df1=pd.DataFrame({'ID':[2,3,4,5],
                 'Name':['A', 'B', 'C', 'D'],
                 'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,6,7,8],
                 'Name':['E', 'F', 'G', 'H'],
                 'Class':[11,23,12,45]})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2, df1])
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
df1=pd.DataFrame({'ID':[2,3,4,5],
                 'Name':['A', 'B', 'C', 'D'],
                 'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[4,5],
                 'Name':['C', 'D'],
                 'Class':[7,8]})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=1, join="inner")
```

```python
df1=pd.DataFrame({'ID':[2,3,4,5],
                 'Name':['A', 'B', 'C', 'D'],
                 'Class':[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({'ID':[1,6,7,8],
                 'Name':['E', 'F', 'G', 'H'],
                 'Class':[11,23,12,45]})
df2
```

```python
pd.concat([df1, df2], keys=['df1', 'df2'])
```

```python
df3=pd.DataFrame({'Marks':[40,50,60,70]})
df3
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], sort=False)
```
